* 输入文件：DDBC_test.csv
* DDBC数据库：DDBC_items_trim.csv
* 所用到的限制条件：时间限制（trim文件朝代为“唐”），名称限制（两者名字要对的上）
* 输出文件包含的内容：输入文件的信息+查找结果匹配的**地名、总ID、规范码、经纬度、行政区（今）、行政级别**

```python
import pandas as pd
```

```python
database = pd.read_csv('DDBC_items_trim.csv',skiprows = 0,encoding='utf-8')
database.head()
```

```python
datasearch = pd.read_excel('DDBC_test.xlsx', encoding='gbk')
datasearch.head()
```

```python
# 精简数据库，只跳出来唐代的数据信息
database_tang=database[database['朝代']=='唐']
database_tang.head()
```

```python
a=database_tang.loc[21].tolist()
a
```

```python
# step 1 建立一个新的dataframe，用来储存处理检索之后的数据，大写字母的attributes储存输入文件的内容，小写字母的attributes储存tgaz检索返回的结果

data = [["ID"],
        ["NAME"],
        ["FIRSTYEAR"],
        ["LASTYEAR"],
        ['DAO'],
        ["count"],
        ["source"],
        ["name_found"],
        ["name_bieming"],
        ["latitude"],
        ["longitude"],
        ["总id"],
        ["行政区"],
        ["行政级别"],
        ["规范码"],
        ["属"]]
new_df=pd.DataFrame(data, columns=["attributes"]).set_index("attributes")
display(new_df)
```

```python
# step 2
# 1）dataframe里面地名和起止年的读取
name_series=datasearch["name_trad"] # 储存地名
year_fy_series=datasearch['time_ft_col'] # 储存开始年份
year_ly_series=datasearch['time_lt_col'] # 储存结束年份
ID_series=datasearch['new_id'] # 储存原ID
dao_series=datasearch['level1_chn'] # 储存所属“道”的信息
print(len(name_series))
```

### 名称对比的原则
对目标文件的【地名】进行清理（删除所有末尾的“州/县/府”之类的文字）并另存为单独的column，用新的column进行string对比，如果目标文件的地名include在DDBC的地名里面，或者DDBC的地名include在目标文件的地名里面，就判断为TRUE，只需要把所有TRUE的结果对应内容保存在表格里面以供之后人工对比即可

```python
# 2）-1 建立检索函数，name是检索的地名，df是DDBC的数据表
def find_placenames(name,df):
    places_list=[]
    for i in df.index.tolist():
        diming=str(df['地名'][i])
        bieming=str(df['別名'][i])
        name_str=str(name)
        if ((name_str in diming) or (diming in name_str) or (name_str in bieming) or (bieming in name_str)):
        # 之前的对比是“name==df['地名'][i]”
            item=df.loc[i].tolist()
            places_list.append(item) #places_list里面储存每一条DDBC里面的地名和检索数据地名匹配的数据的所有信息，后面再拆分放入dataframe

    return(places_list)
```

```python
# 3）-1 定义函数，对每个values里面的内容进行拆解并不断赋给new_dataframe里面相应的地方
def add_values(new_df,places_list,original_ID,original_name,year_fy, year_ly,dao):
    i0=original_ID
    j=0
    if places_list:
        for item in places_list:
            j+=1
            i=str(i0)+"_"+str(j) #用来表示原目标文件每个ID有多少个检索结果
            new_df[i]=None
            new_df[i]["ID"]=original_ID
            new_df[i]["NAME"]=original_name
            new_df[i]["FIRSTYEAR"]=year_fy
            new_df[i]["LASTYEAR"]=year_ly
            new_df[i]["DAO"]=dao
            new_df[i]["count"]=len(places_list)
            new_df[i]["source"]="DDBC"
            new_df[i]["name_found"]=item[2]
            new_df[i]["name_bieming"]=item[3]
            new_df[i]['latitude']=item[7]
            new_df[i]['longitude']=item[8]
            new_df[i]['总id']=item[0]
            new_df[i]["行政区"]=item[9]
            new_df[i]["行政级别"]=item[10]
            new_df[i]["规范码"]=item[4]
            new_df[i]["属"]=item[11]
    else:
        i=str(i0)+"_"+str(j)
        new_df[i]=None
        new_df[i]["ID"]=original_ID
        new_df[i]["NAME"]=original_name
        new_df[i]["FIRSTYEAR"]=year_fy
        new_df[i]["LASTYEAR"]=year_ly
        new_df[i]["DAO"]=dao
        new_df[i]["count"]=0
        new_df[i]["source"]="DDBC"
        new_df[i]["name_found"]=None
        new_df[i]["name_bieming"]=None
        new_df[i]['latitude']=None
        new_df[i]['longitude']=None
        new_df[i]['总id']=None
        new_df[i]["行政区"]=None
        new_df[i]["行政级别"]=None
        new_df[i]["规范码"]=None
        new_df[i]["属"]=None
    return(new_df)
```

```python
# 3）-2 对提取的原表格内容进行遍历循环，调用前面定义的两个函数
# char_list=['府','州','县','军','监'，'使'] 既然用“in”的方式直接进行字符判断，似乎就不需要清除府州军监这些character了
for k in range(0,len(name_series)):
    search_ID=ID_series[k]
    search_name=name_series[k]
    search_year_fy=year_fy_series[k] 
    search_year_ly=year_ly_series[k] 
    search_dao=dao_series[k]
    search_results=find_placenames(search_name,database_tang)
    final_df=add_values(new_df,search_results,search_ID,search_name,search_year_fy, search_year_ly, search_dao)
display(new_df)
```

```python
# 转置dataframe，并且按照ID 的大小进行排序
trans_new_df=new_df.transpose()
final_df=trans_new_df.sort_values("ID")
display(final_df)
```

```python
# DDBC中匹配到的结果
j=0
for i in final_df['name_found']:
    if i!=None:
        j+=1
print(j)
```
